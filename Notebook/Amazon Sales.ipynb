```python
price_cols = ['actual_price', 'discounted_price', 'rating', 'rating_count']

for col in price_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace('₹','', regex=False)
        .str.replace(',','', regex=False)
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['discount_percentage'] = (
    (df['actual_price'] - df['discounted_price']) / df['actual_price']
) * 100
```

```python
df[['actual_price','discounted_price','discount_percentage']].head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
top_categories = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Categories by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='discounted_price',
    y='rating',
    alpha=0.6
)
plt.title("Price vs Product Rating")
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='discount_percentage',
    y='rating',
    alpha=0.6
)
plt.title("Discount Percentage vs Rating")
plt.xlabel("Discount %")
plt.ylabel("Rating")
plt.show()
```

```python
top_products = df.sort_values('rating_count', ascending=False).head(10)

top_products[['product_name','rating','rating_count']]
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(
    df[['actual_price','discounted_price','discount_percentage','rating','rating_count']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Matrix")
plt.show()
```

```python
bins = [0, 500, 1000, 2000, 5000, df['discounted_price'].max()]
labels = ['Budget', 'Low-Mid', 'Mid', 'Premium', 'Luxury']

df['price_segment'] = pd.cut(
    df['discounted_price'],
    bins=bins,
    labels=labels
)
```

```python
df['price_segment'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='price_segment',
    y='rating'
)
plt.title("Ratings Distribution by Price Segment")
plt.xlabel("Price Segment")
plt.ylabel("Rating")
plt.show()
```

```python
discount_bins = [0, 10, 25, 50, 75, 100]
discount_labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']

df['discount_group'] = pd.cut(
    df['discount_percentage'],
    bins=discount_bins,
    labels=discount_labels
)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x='discount_group',
    y='rating',
    ci=None
)
plt.title("Average Rating by Discount Level")
plt.xlabel("Discount Level")
plt.ylabel("Average Rating")
plt.show()
```

```python
review_bins = [0, 50, 200, 1000, df['rating_count'].max()]
review_labels = ['Very Low', 'Low', 'Medium', 'High']

df['review_segment'] = pd.cut(
    df['rating_count'],
    bins=review_bins,
    labels=review_labels
)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='review_segment',
    y='rating'
)
plt.title("Rating Stability vs Review Volume")
plt.xlabel("Review Volume Segment")
plt.ylabel("Rating")
plt.show()
```

```python
category_perf = df.groupby('category').agg(
    avg_price=('discounted_price','mean'),
    avg_rating=('rating','mean'),
    avg_discount=('discount_percentage','mean'),
    total_reviews=('rating_count','sum')
).sort_values('total_reviews', ascending=False)

category_perf.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=category_perf,
    x='avg_price',
    y='avg_rating',
    size='total_reviews',
    sizes=(50,1000),
    alpha=0.7
)
plt.title("Category Performance Map")
plt.xlabel("Average Price")
plt.ylabel("Average Rating")
plt.show()
```

```python
df['value_score'] = (
    df['rating'] * df['discount_percentage']
) / df['discounted_price']
```

```python
top_value = df.sort_values('value_score', ascending=False).head(10)

top_value[['product_name','rating','discount_percentage','discounted_price','value_score']]
```
